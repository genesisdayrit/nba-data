```python
from nba_api.stats.endpoints import commonallplayers
from nba_api.stats.endpoints import playercareerstats
import pandas as pd

# Fetch all players data
all_players = commonallplayers.CommonAllPlayers().get_data_frames()[0]

# Filter for players who are currently active
active_players = all_players[all_players['TO_YEAR'] == '2023'] 

# Initialize an empty DataFrame to hold all player stats
all_player_stats = pd.DataFrame()

# Fetch career stats for each active player
for index, player in active_players.iterrows():
    print(f"Fetching data for {player['PLAYER_NAME']}...")
    career = playercareerstats.PlayerCareerStats(player_id=player['PERSON_ID'])
    df = career.get_data_frames()[0]
    all_player_stats = pd.concat([all_player_stats, df])

# Display the combined stats
all_player_stats
```

```python
from nba_api.stats.endpoints import playergamelog, commonallplayers
import pandas as pd

# Fetch all players data
all_players = commonallplayers.CommonAllPlayers().get_data_frames()[0]

# Filter for players who are currently active
active_players = all_players[all_players['TO_YEAR'] == '2023'] 

# Initialize an empty DataFrame to hold all game logs
all_game_logs = pd.DataFrame()

# Fetch playoff game logs for each active player
for index, player in active_players.iterrows():
    print(f"Fetching playoff data for {player['PLAYER_NAME']}...")
    game_log = playergamelog.PlayerGameLog(player_id=player['PERSON_ID'], season_type_all_star='Playoffs')
    df = game_log.get_data_frames()[0]
    all_game_logs = pd.concat([all_game_logs, df])

# Display the combined game logs
all_game_logs
```

```python
from nba_api.stats.endpoints import playergamelog, commonallplayers
import pandas as pd

# Get a DataFrame with info about all players
all_players_df = commonallplayers.CommonAllPlayers().get_data_frames()[0]

# Initialize an empty DataFrame to hold the game logs
all_game_logs_df = pd.DataFrame()

# Iterate over each player in all_players_df
for index, player in all_players_df.iterrows():
    print(f"Fetching data for {player['DISPLAY_FIRST_LAST']}...")

    # Get the game log for this player
    game_log = playergamelog.PlayerGameLog(player_id=player['PERSON_ID'])
    player_game_log_df = game_log.get_data_frames()[0]

    # Add the player's game log to all_game_logs_df
    all_game_logs_df = pd.concat([all_game_logs_df, player_game_log_df])

# Now, all_game_logs_df contains game logs for all players
```

```python
import time

all_game_logs_df = None

# Iterate through all players
for i, player in all_players_df.iterrows():
    print(f"Fetching data for {player['DISPLAY_FIRST_LAST']}...")
    
    # Get the game log for this player
    game_log = playergamelog.PlayerGameLog(player_id=player['PERSON_ID'])
    player_game_log_df = game_log.get_data_frames()[0]
    
    # Add the player's game log to all_game_logs_df
    if all_game_logs_df is None:
        all_game_logs_df = player_game_log_df
    else:
        all_game_logs_df = pd.concat([all_game_logs_df, player_game_log_df])
    
    # Add a delay before the next request
    time.sleep(2)  # wait for 2 seconds

print("Done fetching all game logs!")
```
